```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('../data/creditcard.csv')

print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
print("Dataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nBasic Statistics:")
df.describe()
```

```python
fraud_count = df['Class'].value_counts()
fraud_percent = df['Class'].value_counts(normalize=True) * 100

print("Transaction Counts:")
print(fraud_count)
print("\nPercentage:")
print(fraud_percent)

# Plot
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df, palette=['green','red'])
plt.title('Legitimate vs Fraud Transactions')
plt.xticks([0,1], ['Legitimate', 'Fraud'])
plt.ylabel('Count')
plt.show()
```

```python
fraud = df[df['Class'] == 1]
legit = df[df['Class'] == 0]

print("Fraud Transaction Amount Stats:")
print(fraud['Amount'].describe())

print("\nLegitimate Transaction Amount Stats:")
print(legit['Amount'].describe())

# Plot
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(legit['Amount'], bins=50, color='green', alpha=0.7)
plt.title('Legitimate Transaction Amounts')
plt.xlabel('Amount')

plt.subplot(1,2,2)
plt.hist(fraud['Amount'], bins=50, color='red', alpha=0.7)
plt.title('Fraud Transaction Amounts')
plt.xlabel('Amount')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.hist(legit['Time'], bins=50, color='green', alpha=0.7)
plt.title('Legitimate - Time Distribution')
plt.xlabel('Time (seconds)')

plt.subplot(1,2,2)
plt.hist(fraud['Time'], bins=50, color='red', alpha=0.7)
plt.title('Fraud - Time Distribution')
plt.xlabel('Time (seconds)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
corr_with_fraud = df.corr()['Class'].drop('Class').sort_values()

plt.figure(figsize=(10,8))
corr_with_fraud.plot(kind='barh', color=['red' if x < 0 else 'green' for x in corr_with_fraud])
plt.title('Feature Correlation with Fraud')
plt.xlabel('Correlation')
plt.axvline(x=0, color='black', linewidth=0.5)
plt.show()

print("\nTop 5 features positively correlated with fraud:")
print(corr_with_fraud.tail(5))

print("\nTop 5 features negatively correlated with fraud:")
print(corr_with_fraud.head(5))
```
